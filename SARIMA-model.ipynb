```python
import random
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Import data
df_train = pd.read_csv('scaled_train.csv', sep=';')
df_test = pd.read_csv('scaled_test.csv', sep=';')
```

```python
# Check for correct import
pd.set_option('display.max_colwidth', 2000)

df_train.head()

# Check for rows and columns size
# print(df_train.shape)
```

```python
# Fit the ARIMA model
y_train = df_train['prec']
x_train = df_train.drop(['prec'], axis=1)
model = ARIMA(y_train, exog=x_train, order=(1, 0, 1), seasonal_order=(1, 0, 0, 365))
model_fit = model.fit()
```
